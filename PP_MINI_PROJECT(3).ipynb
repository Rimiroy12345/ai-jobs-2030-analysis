```python
# creating 3 histograms 
# Y-axis and counts and X-axis as Risk_Category , Education_Level and Total_skills_required
```

```python
import matplotlib.pyplot as p
fig, axes = p.subplots(3, 1, figsize=(8, 8))

# Plot 1: Risk Category
axes[0].hist(df['Risk_Category'])
axes[0].set_title("Risk Category")

# Plot 2: Education Level
axes[1].hist(df['Education_Level'])
axes[1].set_title("Education Level")

# Plot 3: Total Skills Required
axes[2].hist(df['Total_skills_required'])
axes[2].set_title("Total Skills Required")

interpretation = (
    "The histogram of Risk Category shows that most individuals fall into the Medium-risk group "
    "for AI-driven job automation by 2030. The High-risk group is the smallest, suggesting limited "
    "roles face full automation. The Education Level distribution shows that High School and "
    "Bachelorâ€™s degrees are the most common, while PhD holders are few." "The Total Skills Required "
    "histogram reveals that most people possess around 4.5â€“6 skills, indicating moderate skill "
    "diversity. A higher number of skills generally reduces the likelihood of AI replacement "
    "because multi-skill roles involve more complex, non-routine tasks."
)

fig.text(
    0.5, -0.10,
    interpretation,
    wrap=True,
    ha='center',
    fontsize=9
)

p.tight_layout()
p.show()
```

```python
# creating count plot for " Risk Category by Education Level "
# X-axis as Category and Y-axis as Count
```

```python

p.figure(figsize=(8,6))

s.countplot(x="Risk_Category", hue="Education_Level", data=df)
p.title("Risk Category by Education Level")
p.xlabel("Risk Category")
p.ylabel("Count")

interpretation = (
    "This bar chart shows how many jobs at each education level fall into different AI automation "
    "risk categories by 2030. For all four groups (High School, Bachelorâ€™s, Masterâ€™s, PhD), most jobs "
    "are in the Medium-risk category, which means they are likely to be reshaped but not fully automated. "
    "High- and low-risk jobs are fewer and distributed across education levels, suggesting that no "
    "qualification band is completely safe or uniquely exposed to extreme automation risk."
)

p.tight_layout(pad=2)

p.figtext(0.5, -0.15, interpretation, wrap=True, ha='center', fontsize=11)

p.show()
```

```python
# creating heatmap of crosstab for " Relationship: Education Level vs Risk Category "
```

```python
relation_table = pd.crosstab(df["Education_Level"], df["Risk_Category"])
p.figure(figsize=(6,4))
s.heatmap(relation_table, annot=True, cmap="Blues")
p.title("Relationship: Education Level vs Risk Category")
interpretation = (
    "The medium-risk category dominates across all education levels, indicating that most jobs\n"
    "face moderate exposure to AI automation regardless of qualification. High School graduates\n"
    "appear slightly more represented in medium-risk roles. High and low risk categories remain\n"
    "relatively uniform across all groups, implying that education level alone does not strongly\n"
    "determine automation riskâ€”other factors such as job type and skill specialization matter more."
)
p.figtext(0.5, -0.25, interpretation, wrap=True, ha='center', fontsize=9)
p.tight_layout()
p.show()
```

```python
# creating subplots for " Experience by Job Title "
# X-axis as Job Title , Y-axis as Total Years of Experience   Experience by Job Title
```

```python
# creating heatmap for " Correlation Heatmap of Job & Skill Features "
```

```python
df2 = df.drop('Job_Title', axis=1)
df3 = df2.drop('Education_Level', axis=1)
df4 = df3.drop('Risk_Category', axis=1)

p.figure(figsize=(14, 12))

corr = df4.corr()

s.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5,
    linecolor="gray",
    square=True,
    cbar_kws={"shrink": 0.7},
    center=0
)

p.title("Correlation Heatmap of Job & Skill Features", fontsize=16, pad=20)
p.xticks(rotation=45, ha='right', fontsize=10)
p.yticks(rotation=0, fontsize=10)

p.tight_layout()
p.subplots_adjust(bottom=0.32)

interpretation = (
    "Interpretation:\n"
    "1. This heatmap shows Pearson correlation coefficients (-1 to 1) for job-related factors and skills.\n"
    "2. Total_skills_required has a strong positive relationship with individual skills (Skill 2, 3, 4, 10, etc.).\n"
    "3. Average_Salary, Years_Experience, AI_Exposure_Index, and Tech_Growth_Factor\n"
    "   show weak or near-zero correlations with specific skills.\n"
    "4. The only noticeable non-skill correlation is between Average_Salary and Years_Experience,\n"
    "   though it remains low.\n"
    "5. The key takeaway: Skills are independent of salary, experience, or AI exposure,\n"
    "   but they collectively shape Total_skills_required."
)

p.figtext(
    0.01,
    0.02,
    interpretation,
    ha='left',
    fontsize=12
)

p.show()
```

```python
# creating bar grapgh for " Skill Scores for First 5 Jobs "
# X-axis as Skills , Y-axis as Skill Scores and legend as jobs
```

```python
import matplotlib.pyplot as p
import numpy as np

# Data
jobs = ['Security Guard', 'Research Scientist', 'Construction Worker', 'Software Engineer', 'Financial Analyst']
skill_labels = ['Skill_1', 'Skill_2', 'Skill_3', 'Skill_4', 'Skill_5']
skills = {
    'Security Guard': [0.45, 0.1, 0.46, 0.33, 0.14],
    'Research Scientist': [0.02, 0.52, 0.4, 0.05, 0.97],
    'Construction Worker': [0.01, 0.94, 0.56, 0.39, 0.02],
    'Software Engineer': [0.43, 0.21, 0.57, 0.03, 0.84],
    'Financial Analyst': [0.75, 0.54, 0.59, 0.97, 0.61]
}

# Set bar width and positions
bar_width = 0.15
x = np.arange(len(skill_labels))

# Create bars for each job
for i, job in enumerate(jobs):
    p.bar(x + i*bar_width, skills[job], width=bar_width, label=job)

# Configure axes
p.xlabel('Skills')
p.ylabel('Skill Scores')
p.title('Skill Scores for First 5 Jobs')
p.xticks(x + bar_width*2, skill_labels)
p.legend()

# Show plot
p.show()
```
